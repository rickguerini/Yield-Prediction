```python
import numpy as np 
import pandas as pd 
```

```python
GeoChemData = pd.read_csv('./GC_Data_relim.csv')
GeoChemData.head()
```

```python
GeoChemData.dtypes
```

```python
# drop date columns
GeoChemData = GeoChemData.drop('full_date', 1)
```

```python
GeoChemData.describe()
```

```python
# convert yield to percentage
GeoChemData['Yield'] = GeoChemData['Yield']*100
```

```python
# remove missing data in yield
GeoChemData = GeoChemData.dropna()
```

```python
# separate into features and label
X = GeoChemData.loc[:, GeoChemData.columns != 'Yield']
y = GeoChemData['Yield']
GeoChemData.dtypes
```

```python
import seaborn as sns
transpose = GeoChemData.transpose()
correlation_matrix = np.corrcoef(transpose)
sns.heatmap(correlation_matrix)
```

```python
from sklearn import preprocessing, pipeline, linear_model
# Make up our pipeline where we transform the geochem data to make it more gaussian
regressor = pipeline.Pipeline([
    ('transform', preprocessing.FunctionTransformer(np.log, validate=True)),
    ('model', linear_model.LinearRegression())
])
```

```python
from sklearn.model_selection import train_test_split #split arrays or matrices into random train and test subsets

# split data train 70% and test 30%. You can try other splits here.
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, 
                                                    random_state=42)
```

```python
regressor.fit(x_train, y_train)
```

```python
prediction = regressor.predict(x_test)
```

```python
regressor.score(x_test, y_test)
```

```python
from sklearn import tree
regressor = pipeline.Pipeline([
    ('transform', preprocessing.FunctionTransformer(np.log, validate=True)),
    ('model', tree.DecisionTreeRegressor())
])
```

```python
regressor.fit(x_train, y_train)
```

```python
regressor.score(x_test, y_test)
```
